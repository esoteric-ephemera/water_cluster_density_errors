```python
import pandas

from database_tools import (
    get_geometry_for_single_dataset,
    get_reference_energies_by_dataset,
    get_total_energies_by_dataset,
    get_energies_and_errors_by_dataset_and_functional,
    water_cluster_datasets
)

methods = ["SCAN@HF","HF-r2SCAN-DC4","r2SCAN","r2SCAN50"]
data = {
    k: get_energies_and_errors_by_dataset_and_functional(k,functional=methods)
    for k in water_cluster_datasets
}

table = []
for method in methods:
    table.append([
        data["BEGDB_H2O"][method][f"{k}_energy"]["statistics"]["mean_absolute_deviation"]
        for k in ("binding","interaction","relaxation")
    ] + [
        data[k][method]["binding_energy"]["statistics"]["mean_absolute_deviation"]
        for k in water_cluster_datasets if k != "BEGDB_H2O"
    ])
table = pandas.DataFrame(
    table,
    columns = [f"BEGDB {k} MAD" for k in ("BE","IE","RE") ] + [f"{k} MAD" for k in ("WATER27","H2O-alkali", "H2O-halide")],
    index = methods
)
table
```
